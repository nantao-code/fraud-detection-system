```python
import pandas as pd
import numpy as np

data = pd.read_csv('./test_creditcard.csv')
data.head()
```

```python
data['label'] = np.random.randint(0, 300, size=len(data))
data
```

```python
data.to_csv('./data_resression.csv', index=False)
```

```python
# 分层采样（按类别比例采样）
fraud_df = data[data['Class'] == 1]
normal_df = data[data['Class'] == 0]

# 从正常交易中采样与欺诈交易相同数量的样本
normal_sample = normal_df.sample(n=10000, random_state=42)

# 合并平衡数据集
balanced_df = pd.concat([fraud_df, normal_sample])
balanced_df
```

```python
balanced_df = balanced_df.reset_index()
balanced_df
```

```python
balanced_df.drop(['index'],inplace=True,axis=1)
balanced_df
```

```python
balanced_df.to_csv('./test_cesditcard.csv', index=False)
```
